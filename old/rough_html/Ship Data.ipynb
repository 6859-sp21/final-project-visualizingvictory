```python
df_list = [A_ships,G_ships,J_ships,M_ships,S_ships]
alphabetical_ships = pd.concat(df_list,ignore_index=True)
```

```python
ship_dict = dict(zip(alphabetical_ships["MC hull no."].to_numpy().astype(str),alphabetical_ships[["Namesake","Fate"]].to_numpy()))
```

```python
all_ship_props = np.array([ship_dict.get(str(hull_no),["null","null"]) for hull_no in all_ships["Hull_Number"]])
```

```python
all_ship_props
```

```python
all_ships["Namesake"] = all_ship_props[:,0]
all_ships["Fate"] = all_ship_props[:,1]
all_ships.head()
```

```python
all_ships_ordered = all_ships.copy().sort_values("Completed")
all_ships_ordered = all_ships_ordered.iloc[:-1].copy()
all_ships_ordered.iloc[-1]
```

```python
fig,ax = plt.subplots(dpi=240)
plt.plot_date(all_ships_ordered["Completed"],np.arange(1,2711),"r-",xdate=True,)
```
